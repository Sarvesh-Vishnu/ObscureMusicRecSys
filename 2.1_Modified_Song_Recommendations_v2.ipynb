## Generating token for recommendation

```python
import requests
import base64
```

```python
def get_token(clientId,clientSecret):
    url = "https://accounts.spotify.com/api/token"
    headers = {}
    data = {}
    message = f"{clientId}:{clientSecret}"
    messageBytes = message.encode('ascii')
    base64Bytes = base64.b64encode(messageBytes)
    base64Message = base64Bytes.decode('ascii')
    headers['Authorization'] = "Basic " + base64Message
    data['grant_type'] = "client_credentials"
    r = requests.post(url, headers=headers, data=data)
    token = r.json()['access_token']
    return token
```

## Track based recommendations

```python
import requests
import random
import pandas as pd
import matplotlib.pyplot as plt

def get_track_recommendations(seed_tracks, token, genres=None, target_popularity=random.randint(1,50)):
    limit = 10
    recUrl = f"https://api.spotify.com/v1/recommendations?limit={limit}&seed_tracks={seed_tracks}"
    
    if genres:
        recUrl += f"&seed_genres={','.join(genres)}"
    
    recUrl += f"&target_popularity={target_popularity}"

    headers = {
        "Authorization": "Bearer " + token
    }

    res = requests.get(url=recUrl, headers=headers)
    if res.status_code != 200:
        raise Exception(f"Spotify API Error: {res.json()}")
    return res.json()
```

```python
# Example usage with expanded genre selection and lower popularity target
genres = ['hip-hop', 'classical', 'k-pop', 'latin', 'indie']
json_response = get_track_recommendations(song_id, token)
```
